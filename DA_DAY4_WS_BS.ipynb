```python
#extracting a table from a webpage
rm="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
wpage=requests.get(rm)
```

```python
code1=BeautifulSoup(wpage.text,'html')
print(code1)
```

```python
tab=code1.find("table")
print(tab)
```

```python
h=tab.find_all('th')
print(h)
```

```python
heading=[i.text for i in h]
print(heading)
```

```python

import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
rows=tab.find_all("tr")
print(rows)
```

```python
data=tab.find_all("td")
print(data)
```

```python
data_act=[i.text for i in data]
print(data_act)
```

```python
data_act[0:4]
```

```python
dataframe=pd.DataFrame(columns=heading)
print(dataframe)
```

```python
st=0
sp=4
for i in range(len(data_act)//4):
  mydf.loc[i]=data_act[st:sp]
  st=st+4
  sp=sp+4
mydf
```

```python
ind=0
st=0
sp=4
for i in data_act:
  if sp<=40:
    mydf.loc[ind]=data_act[st:sp]
    st=st+4
    sp=sp+4
    ind=ind+1
print(mydf)
```
