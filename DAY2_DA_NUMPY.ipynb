```python
#array operations
A=np.array([34,56,78,90])#list will throw an error while arrays iterate themseleves
B=np.array([7,9,3,12])
C=A-B
print(A-B)
print(A+B)
```

```python
#min max
A=np.array([34,56,78,90])
print(A.max())#max and min accepts numeric datatypes int and float
print(A.min())
```

```python
#one on one multiplication
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print("_____________________")
print(B)
C=A*B#one on one multiplication
print("_____________________")
print(C)
```

```python
#matrix multiplication
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print("_____________________")
print(B)
C=A@B#matrix multiplication
print("_____________________")
print(C)
```

```python
#matrix multiplication
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print("_____________________")
print(B)
C=A.dot(B)#matrix multiplication
print("_____________________")
print(C)
```

```python
A=np.array([34,56,78,90]).reshape(2,2)
print(A)
print("_____________________")
print(A.max(axis=0))#axis=0 : column
print("_____________________")
print(A.max(axis=1))#axis=1 : row
print("_____________________")
print(A.min(axis=0))#axis=0 : column
print("_____________________")
print(A.min(axis=1))#axis=1 : row
```

```python
A=np.array([34,56,78,90]).reshape(2,2)
B=np.array([7,9,3,12]).reshape(2,2)
print(A)
print("_____________________")
print(B)
print("after vertically stacking arrays!")
print(np.vstack((A,B)))
print("_____________________")
print(np.hstack((A,B)))
print("_____________________")
print(np.stack((A,B),axis=0))
print("_____________________")
print(np.stack((A,B),axis=1))
```

```python
A=np.arange(30).reshape(2,3,5)
print(A)
print("after dstack")
print(np.dstack(A))#no.of rows =no.of groups
#no.of col in every group = no.of rows in every group
[0 0],[1,1],............
#original 2,3,5
#dstack   3,5,2
```

```python
#splitting arrays!
A=np.arange(18).reshape(6,3)
print(A)
print("after splitting")
print(np.vsplit(A,3))#based on rows vsplit happens(no.of rows should be divisble by that number)
print("_____________________")
print(np.hsplit(A,3))
```

```python
A=np.arange(18).reshape(6,3)
print(A)
print("after splitting")
print(np.vsplit(A,(2,3)))#splitting the array using indexes
print("_____________________")
print(np.hsplit(A,(2,3)))
```
