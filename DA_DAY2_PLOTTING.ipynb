```python
#There are 5 subjects psychology,nano technology,medicine,oncology,anthropology 2 students have applied to do masters in any of these 5 fields with the line plot show in which feild they can excel the cutoff for each subject is 70
#st1=[67,45,73,71,57]
#st2=[43,67,68,53,85]
sub=["pyscho","nt","med","onco","anthro"]
st1=[67,45,73,71,57]
st2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student1
plt.plot(sub,st1,label="student1",color="y")
#student2
plt.plot(sub,st2,label="student2",color="black",marker="p",)
#cutoff
plt.plot(sub,cut,label="cutoff",color="r",linestyle=":",linewidth=4)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()

```

```python
#an ind VS aus match is on live,ind has given 298 as target for aus the runs for every 5 overs is given as follows
#ind=[25,51,84,131,160,189,220,250,267,297]
#aus have completed 25 overs
#aus=[15,41,94,110,151]
#plot overs VS runs for both the countries with the india as blue and aus as yellow line give approriate labels and legend the marker for ind is flag and the marker for aus is plus
#take overs from arange fun create runs as array

import numpy as np
IND=np.array([25,51,84,131,160,189,220,250,267,297])
AUS=np.array([15,41,94,110,151])
IND_over=np.arange(5,51,5)
AUS_over=np.arange(5,26,5)
plt.plot(AUS_over,AUS,color="y",label="AUS",marker="+",linewidth=4)
plt.plot(IND_over,IND,color="b",label="INDIA",marker=">",linewidth=4)
plt.xlabel("overs")
plt.ylabel("runs")
plt.title("IND VS AUS")
plt.grid()
plt.legend()
```
