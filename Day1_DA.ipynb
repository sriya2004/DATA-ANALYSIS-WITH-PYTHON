```python
#sets
s={'s',1,'r',2,True}
print(s)
#len
print(len(s))
#type
print(type(s))
#update
s.update("i")
print(s)
#remove
s.remove('i')
print(s)
#pop
s.pop()
print(s)
#clear
s.clear()
print(s)

```

```python
#set operations
s1={1,2,3,4,5}
s2={4,5,6,7,8}
#union
print(s1.union(s2))
#intersection
print(s1.intersection(s2))
#difference
print(s1.difference(s2))
#symmetric_difference
print(s1.symmetric_difference(s2))
#subset
print(s1.issubset(s2))
#superset
print(s1.issuperset(s2))
#disjoint
print(s1.isdisjoint(s2))
```
